<h4>ii. Four users & which sites they (each) visited more than 10 times</h4> 
Between 2019-02-03 00:00:00 and 2019-02-04 23:59:59, there are four users who visited a certain site more than 10 times. Find these four users & which sites they (each) visited more than 10 times. (Simply provides four triples in the form (user_id, site_id, number of visits) in the box below.)

```python
start = time.strptime("2019-02-03 00:00:00", "%Y-%m-%d %H:%M:%S")
end = time.strptime("2019-02-04 23:59:59", "%Y-%m-%d %H:%M:%S")
time_window = User_data[(User_data["ts"]>=start) & (User_data["ts"]<=end)]
grouped = time_window.groupby(["user_id","site_id"]).count()
grouped[grouped["ts"]>=10]
```

<h4>iii. Top three last visit sites</h4> 
For each site, compute the unique number of users whose last visit (found in the original data set) was to that site. For instance, user "LC3561"'s last visit is to "N0OTG" based on timestamp data. Based on this measure, what are top three sites? (hint: site "3POLC" is ranked at 5th with 28 users whose last visit in the data set was to 3POLC; simply provide three pairs in the form (site_id, number of users).)

```python
user_grouped = User_data.groupby("user_id")
```

```python
def getFirstVisit(user,group):
    group = group.sort_values("ts",ascending=True)
    return group.iloc[0,3]
def getLastVisit(user,group):
    group = group.sort_values("ts",ascending=False)
    return group.iloc[0,3]
```

```python
first_sites = []
last_sites = []
users = []
for user, group in user_grouped:
    users.append(user)
    first_sites.append(getFirstVisit(user,group))
    last_sites.append(getLastVisit(user,group))
result = pd.DataFrame([pd.Series(users),pd.Series(first_sites),pd.Series(last_sites)],index=["users","first visit site", "last visit site"]).T
```

```python
result.groupby("last visit site").count().sort_values("users",ascending=False)
```

The answer is ("5NPAU", 992), ("N0OTG", 561), ("QGO3G", 289)

<h3>4. Part 2 Regression model</h3>

<p>The data contains 300 rows and 3 columns (from the left, A, B, and C). Please build a good regression model which explains column C by a function of A and B. 
<p>Since column A, B and C are all numerical values, I used linear regresion with its loss fuction as MSE here.

```python
import numpy as np
import pandas as pd
import scipy
from scipy import optimize
import warnings
warnings.filterwarnings("ignore")
data = pd.read_excel("Adops & Data Scientist Sample Data.xlsx", sheet_name="Q2 Regression",header=None)
data.columns = ["A","B","C"]
data.shape
```

<h4>i. Detect Null values and outliers.</h4>
<p>There's no Null values in our dataset.
<p>My criterion for detecting outlier (high leverage point) is: outside the interval of mean ± (2.33*standard deviation).
<p>If we assume all the 3 variables has enough samples(Central Limit Theorem), 1% of the samples of each variable would be considered as outliers. The threshold could also be adjusted.

```python
data = data.dropna()
```

```python
def outlierDetector(dataseries):
    mean = dataseries.mean()
    std = dataseries.std()
    lower_bond = mean - (2.33*std)
    upper_bond = mean + (2.33*std)
    outliers = []
    for record in dataseries:
        if record < lower_bond or record > upper_bond:
            outliers.append(record)
    return outliers
```

```python
data = data[~data["A"].isin(outlierDetector(data["A"])) & ~data["B"].isin(outlierDetector(data["B"])) & ~data["C"].isin(outlierDetector(data["C"]))]
data.shape
```

<h4>ii. Do the linear regression</h4>
We have our dependant variable with 2 independant variables, the result of our regression should looks like a plane in the space.

```python
def MSE(b):
    x0 = np.array(data["A"])
    x1 = np.array(data["B"])
    y =  np.array(data["C"])
    pred =b[0] + b[1]*x0 + b[2]*x1
    mse = sum(list(map(lambda x: x**2, y - pred)))
    return mse
```

```python
result = scipy.optimize.minimize(MSE, [1,1,1])
result.x
```

```python
print("The relationship between A, B and C is: C = % 5.2f + (% 5.2f * A) + (% 5.2f * B)"%tuple(result.x))
```

<h4>iii. Visualization</h4>

```python
X, Y = np.meshgrid(data["A"], data["B"])
Z = result.x[0] + result.x[1]*X +result.x[2]*Y
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
fig = plt.figure(figsize=[10,10])
ax = Axes3D(fig)
ax.scatter(data["A"],data["B"],data["C"],c="y")
ax.plot_surface(X,Y,Z, cmap="Blues")
```
